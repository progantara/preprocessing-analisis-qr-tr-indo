```python
import os
```

```python
path = "data/juz30/"
folders = ["hannan", "muzammil", "ummi"]
id_reciter = ["07", "08", "09"] # hannan, muzammil, ummi
new_folder = "data/juz30/wavs"

# create the new folder
if not os.path.exists(new_folder):
    os.makedirs(new_folder)

# get all files in the folder
for folder in folders:
    files = os.listdir(os.path.join(path, folder))
    for file in files:
        # get the file name without extension
        file_name = os.path.splitext(file)[0]
        # get the surah number and ayah number
        surah, ayah = file_name.split("-")
        # get the reciter id
        reciter_id = id_reciter[folders.index(folder)]
        # create the new file name
        new_file_name = f"{surah.zfill(3)}{ayah.zfill(3)}_{reciter_id}.wav"
        print(f"File {file} will be copied to {new_file_name}")
        # copy the file to the new folder
        os.system(f"cp {os.path.join(path, folder, file)} {os.path.join(new_folder, new_file_name)}")
        print(f"File {file} copied to {new_file_name}")
```
