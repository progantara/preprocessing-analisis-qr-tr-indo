```python
import librosa
import numpy as np
import os
```

```python
# open all files in data/audio
for filename in os.listdir('data/audio'):
    if filename.endswith('.wav'):
        # load the file
        audio, sr = librosa.load('data/audio/' + filename)
        # save converted file
        np.save('data/audio/' + filename[:-4] + '.wav', audio)

    else:
        print('Skipping ' + filename)
```
