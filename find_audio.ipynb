```python
filename = "data/quran-simple.csv"

# buka file
with open(filename) as file:
    data = []
    # baca baris berikutnya
    for line in file:
        # pisahkan data berdasarkan koma
        row = line.strip().split("|")
        # tambahkan data ke list
        data.append(row)

# buat dictionary kosong untuk menampung jumlah kata pada setiap ayat
jumlah_kata = {}

# iterasi setiap baris data
ayat=1
for row in data:
    # ambil jumlah kata
    jumlah_kata_ayat = len(row[2].split())
    # simpan jumlah kata pada dictionary
    jumlah_kata[ayat] = jumlah_kata_ayat
    ayat+=1
print(len(data))
# urutkan ayat berdasarkan jumlah kata
sorted_ayat = sorted(jumlah_kata.items(), key=lambda x: x[1])

# tampilkan jumlah kata pada setiap ayat
print("Jumlah kata pada setiap ayat:")
for key, value in sorted_ayat:
    print(f"Ayat {key}: {value} kata")

# calculate the average number of words in each verse
average = sum(jumlah_kata.values()) / len(jumlah_kata)
print(f"Rata-rata jumlah kata pada setiap ayat adalah {average:.2f} kata")

# ada berapa ayat yang jumlah katanya lebih dari rata-rata
count = 0
for key, value in jumlah_kata.items():
    if key not in ["terpendek", "terpanjang"]:
        if value > average:
            count += 1
print(f"Ada {count} ayat yang jumlah katanya lebih dari rata-rata")

# ada berapa ayat yang jumlah katanya kurang dari rata-rata
count = 0
for key, value in jumlah_kata.items():
    if key not in ["terpendek", "terpanjang"]:
        if value < average:
            count += 1
print(f"Ada {count} ayat yang jumlah katanya kurang dari rata-rata")
```

```python
# hitung estimasi rataan menggunakan distribusi normal
import math
std_dev = math.sqrt(sum((x - average) ** 2 for x in jumlah_kata.values()) / len(jumlah_kata))
print(f"Standar deviasi: {std_dev:.2f}")

# hitung z-score
z_score = (jumlah_kata[1] - average) / std_dev
print(f"Z-score untuk ayat pertama: {z_score:.2f}")

# hitung z-score untuk setiap ayat
z_scores = {}
for key, value in jumlah_kata.items():
    z_score = (value - average) / std_dev
    z_scores[key] = z_score
    
# urutkan ayat berdasarkan z-score
sorted_z_scores = sorted(z_scores.items(), key=lambda x: x[1])

# tampilkan z-score untuk setiap ayat
print("Z-score untuk setiap ayat:")
for key, value in sorted_z_scores:
    print(f"Ayat {key}: {value:.2f}")
    
# tampilkan ayat dengan z-score terendah
print(f"Ayat dengan z-score terendah: {sorted_z_scores[0][0]}")
# tampilkan ayat dengan z-score tertinggi
print(f"Ayat dengan z-score tertinggi: {sorted_z_scores[-1][0]}")
```

```python
# estimasi rata-rata jumlah kata
import scipy.stats as stats
confidence = 0.95
n = len(jumlah_kata)
t = stats.t.ppf((1 + confidence) / 2, n - 1)
margin_of_error = t * std_dev / math.sqrt(n)
lower_bound = average - margin_of_error
upper_bound = average + margin_of_error
print(f"Estimasi rata-rata jumlah kata: {average:.2f} kata")
```
